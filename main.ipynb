# `Step 2. Prepare the data`

## The number of calls made and minutes used per month

```python
calls_per_month_df = calls_df.copy()
calls_per_month_df['month'] = calls_per_month_df['call_date'].dt.month
calls_per_month_df = calls_per_month_df.groupby(['user_id', 'month'])['duration'].agg(['count', 'sum']).reset_index()
calls_per_month_df = calls_per_month_df.rename(columns={'count': 'calls_count', 'sum': 'total_duration'})

calls_per_month_df['total_duration'] = calls_per_month_df['total_duration'].apply(np.ceil).astype(int)
calls_per_month_df
```

#### <li>Create new table to demostrate:
<li>Number of calls per month for every user</li>
<li>Duration (in minutes) of calls per month for every user</li>
</ul>
</li>    

## The number of text messages sent per month

```python
messages_per_month_df = messages_df.copy()
messages_per_month_df['month'] = messages_per_month_df['message_date'].dt.month
messages_per_month_df = messages_per_month_df.groupby(['user_id', 'month'])['id'].count().reset_index()
messages_per_month_df = messages_per_month_df.rename(columns={'id': 'messages_count'})
messages_per_month_df
```

#### <li>Create new table to demostrate:
<li>Number of messages per month for every user</li>
</ul>
</li>    

## The volume of data per month

```python
internet_per_month_df = internet_df.copy()
internet_per_month_df['month'] = internet_per_month_df['session_date'].dt.month
internet_per_month_df = internet_per_month_df.groupby(['user_id', 'month'])['mb_used'].sum().reset_index()
internet_per_month_df = internet_per_month_df.rename(columns={'mb_used': 'total_mb_used'})
internet_per_month_df['total_mb_used'] = internet_per_month_df['total_mb_used'].apply(np.ceil).astype(int)
internet_per_month_df.head()
```

#### <li>Create new table to demostrate:
<li>Total quantity (in megabites) of Internet data used by users per month</li>
</ul>
</li>    

```python
data_per_month_df = pd.merge(calls_per_month_df, messages_per_month_df,  how='outer', left_on=['user_id','month'], right_on = ['user_id','month'])
data_per_month_df['messages_count'] = data_per_month_df['messages_count'].fillna(0).astype(int)
data_per_month_df['calls_count'] = data_per_month_df['calls_count'].fillna(0).astype(int)
data_per_month_df.head()
```

```python
data_per_month_df = pd.merge(data_per_month_df, internet_per_month_df,  how='outer', left_on=['user_id', 'month'], right_on = ['user_id','month'])
data_per_month_df.head()
```

#### <li>Create new table:
<li>Table combines all three tables above and demonstrate:</li>
<li>Calls, messages, and Internet data used per month for every user</li>
</ul>
</li>    

```python
users_tariffs = users_df[['tariff', 'user_id']]

data_per_month_df = pd.merge(data_per_month_df, users_tariffs, left_on=['user_id'], right_on = ['user_id'], how="inner")

data_per_month_df = data_per_month_df.rename(columns={'tariff': 'plan_name'})
data_per_month_df.head()
```

#### <li>Create new table to demostrate:
<li>Tariff plan's name linked to every user</li>

</ul>
</li>    

```python
data_per_month_df = pd.merge(data_per_month_df, plans_df,  how='outer', left_on=['plan_name'], right_on = ['plan_name'])
data_per_month_df['messages_count'] = data_per_month_df['messages_count'].fillna(0).astype(int)
data_per_month_df['calls_count'] = data_per_month_df['calls_count'].fillna(0).astype(int)
data_per_month_df['messages_count'] = data_per_month_df['messages_count'].fillna(0).astype(int)
data_per_month_df.head()
```

## The monthly profit from each user 

```python
monthly_profit_df = data_per_month_df.copy()

monthly_profit_df['calls_min_used'] = monthly_profit_df['total_duration'] - monthly_profit_df['minutes_included']
monthly_profit_df.loc[monthly_profit_df['calls_min_used'] < 0, 'calls_min_used'] = 0
monthly_profit_df['calls_min_used'] = monthly_profit_df['calls_min_used'].apply(np.ceil).fillna(0).astype(int)

monthly_profit_df['messages_used'] = monthly_profit_df['messages_count'] - monthly_profit_df['messages_included']
monthly_profit_df.loc[monthly_profit_df['messages_used'] < 0, 'messages_used'] = 0

monthly_profit_df['internet_gb_used'] = monthly_profit_df['total_mb_used'] - monthly_profit_df['mb_per_month_included']
monthly_profit_df.loc[monthly_profit_df['internet_gb_used'] < 0, 'internet_gb_used'] = 0
monthly_profit_df['internet_gb_used'] = monthly_profit_df['internet_gb_used'] / 1024
monthly_profit_df['internet_gb_used'] = monthly_profit_df['internet_gb_used'].apply(np.ceil).fillna(0).astype(int)

monthly_profit_df.head()
```

#### <li> Add 3 new columns which demonstrate:
<li>Whether or not user exceed prepaid plan</li>

</ul>
</li>    

```python
monthly_profit_df['extra_calls_cost'] = monthly_profit_df['calls_min_used'] * monthly_profit_df['usd_per_minute']
monthly_profit_df['extra_messages_cost'] = monthly_profit_df['messages_used'] * monthly_profit_df['usd_per_message']
monthly_profit_df['extra_internet_cost'] = monthly_profit_df['internet_gb_used'] * monthly_profit_df['usd_per_gb']

monthly_profit_df.head()
```

#### <li> Add 3 new columns which demonstrate:
<li>Additional payments for calls, messages, and internet data in case user exceeded plan limits</li>

</ul>
</li>    

```python

def final_payment (row):
    if row['plan_name'] == 'surf':
        return row['extra_messages_cost'] + row['extra_internet_cost'] + row['extra_calls_cost'] + 20
    if row['plan_name'] == 'ultimate':
        return row['extra_messages_cost'] + row['extra_internet_cost'] + row['extra_calls_cost'] + 70
monthly_profit_df['user_payment'] = (
    monthly_profit_df['usd_monthly_pay'] +
    monthly_profit_df['extra_calls_cost'] +
    monthly_profit_df['extra_messages_cost'] +
    monthly_profit_df['extra_internet_cost']
)

monthly_profit_df.info()
monthly_profit_df
```

#### <li> Add column which demonstrate:
<li>Monthly payment of each user</li>

</ul>
</li>

# `Step 4. Test the hypotheses`

```python
#The average profit from users of Ultimate and Surf calling plans differs

alpha = 0.05

results = st.ttest_ind(
        ultimate_plan_df['user_payment'], 
        surf_plan_df['user_payment'])
print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("We can't reject the null hypothesis")
else:
    print("We reject the null hypothesis")
    
#  p-value: 9.470478215100305e-09
#We can't reject the null hypothesis  
```

#### <li> We can't reject the null hypothesis:
<li>It supports our data from calculations above which shows that<br>
Surf paln is almost twice more profitable than Ultimate
</ul>
</li>

####  Null hypotheses: Since the cost of both prepaid plans differ, we believe that total profit differs as well.
#### Alternative hypothesis: Since the cost of Ultimate plan is almost 3 times higher than that of Surf, we believe taht users tend to choose the one which is cheaper. Thus, it makes final totals from both plans to be similar.


```python
users_payments_df = final_payment_df.copy()

payment_per_state_df = pd.merge(users_payments_df, users_df,  how='inner', left_on=['user_id'], right_on = ['user_id'])
payment_per_state_df
```

#### <li> Two tables contain data of users From NY-NJ and Other states:

</ul>
</li>

```python
ny_nj_df = payment_per_state_df[payment_per_state_df['city'] == 'New York-Newark-Jersey City, NY-NJ-PA MSA']

ny_nj_df
```

```python
other_states_df = payment_per_state_df[payment_per_state_df['city'] != 'New York-Newark-Jersey City, NY-NJ-PA MSA']
other_states_df
```

```python
#The average profit from users in NY-NJ area is different from that of the users from other regions.
alpha = .05

results = st.ttest_ind(
        ny_nj_df['user_payment'], 
        other_states_df['user_payment'])
print('p-value:', results.pvalue)

if (results.pvalue < alpha):
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
    
#p-value: 0.04076784279879048
#We reject the null hypothesis    
```

####  Null hypotheses: Since the New York and New Jarsey are the most rich cities in United States, that makes us believe users from this area give more profit than those from other states.
#### Alternative hypothesis: Despite New York and New Jarsey's population financial well-being, avarage profit from users in this area and other regions are not different because people in average use the same amount of calls, messages, and data.

#### <li> We reject the null hypothesis :
<li>It seems that particular location (state) doesn't make any diferece for amount of profit</li>

</ul>
